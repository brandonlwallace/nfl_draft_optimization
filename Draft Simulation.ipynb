This notebook takes scaped data of hundreds of 2023 NFL mock drafts from sports journalists to build a drafting tool. NFL teams typically enter the NFL draft with several college football players that they want to join their team. However, the draft is a highly uncertain enviorment. No team has perfect information about the intentions of other teams. Since the draft is ordered, the behavior of the teams preceeding another is studied acutely. 

This project finds a distribution for all players in the draft based on how often they are mocked at each pick. I use large simulations to investigate key questions facing a General Manager. Lastly, I focus on the Pittsburgh Steelers, my favorite team. I find the likelihood that a player the Steelers want is available at their pick location, 17. 

```python
import pandas as pd
import numpy as np
import random
import warnings
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
warnings.filterwarnings('ignore')
```

```python
# Load Excel file into Pandas dataframe
column_names =  ["Publication", "Author", "Date"] + [f'pick {i}' for i in range(1, 43)]
df = pd.read_csv('mock_draft_data.csv', header=None)
df.columns = column_names

# Clean data 
df['Publication'] = df['Publication'].astype(str).str.replace(r"\[|\]|'", '')
df['Author'] = df['Author'].astype(str).str.extract(r"By (\w+ \w+)")[0]
df['Date'] = df['Date'].astype(str).str.replace(r"\[|\]|'", '')
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y') 
# Remove non-first round picks
df = df.iloc[:, :-10]
```

```python
# Checking Cleaned Data
df.head(5)
```

```python
# Extract player names from pick columns
pick_columns = df.columns[3:]  # Assumes pick columns start from index 3

# Create Melted df and distribution of picks
melted_df = df.melt(id_vars=['Publication', 'Author', 'Date'], value_vars=pick_columns, var_name='Pick', value_name='Player')

# Get full distribtuion 
distribution = melted_df.groupby(['Player', 'Pick'])['Player'].count().sort_values(ascending=False)
```

## Simulations

#### How does a simulation predict the actual draft will go?

```python
# Transform distribution into dataframe
p_p_df = distribution.to_frame(name='Count').reset_index()
p_p_df = p_p_df.sort_values('Count', ascending=False).reset_index(drop=True)
```

```python
p_p_df.head()
```

```python
# Single Simulated Draft based on Mock Draft Data 

draft_results = []
selected_players = set()
players_list = p_p_df['Player'].tolist()

for i in range(1, 33):
    available_players = p_p_df[p_p_df['Pick'] == f'pick {i}']
    available_players = available_players[~available_players['Player'].isin(selected_players)]
    
    if len(available_players) > 0:
        available_players = available_players.reset_index(drop=True)
        
        # Create a list of players weighted by the count column
        weighted_players = []
        for _, row in available_players.iterrows():
            player = row['Player']
            count = row['Count']
            weighted_players.extend([player] * count)
        
        selected_player = random.choice(weighted_players)
        draft_results.append((selected_player, f'pick {i}'))
        selected_players.add(selected_player)

'''
The new list called weighted_players is created to store the players weighted by the count column. 
For each player in the available players, the player's name is repeated in weighted_players based on its count. 
This creates a weighted distribution where players with higher counts have a greater chance of being selected.
The random.choice() function is then applied to weighted_players to select a player based on the weights. 
This ensures that players with higher counts are more likely to be chosen during the simulation.
By incorporating the count column as a weight in the player selection process, 
the simulation will consider the frequency of player selection at each pick based on mock drafts and make selections accordingly.
'''

print("NFL Draft First Round Simulation:")
for i, result in enumerate(draft_results, start=1):
    player, pick = result
    print(f"Pick {i}: {player}")
```

```python
# Monte Carlo Simulation of 1,000 drafts 

# Number of simulations
num_simulations = 1000

# List to store the results of each simulation
all_results = []

# Repeat the same simulation for the specified number of times
for _ in range(num_simulations):
    draft_results = []
    selected_players = set()
    players_list = p_p_df['Player'].tolist()

    for i in range(1, 33):
        available_players = p_p_df[p_p_df['Pick'] == f'pick {i}']
        available_players = available_players[~available_players['Player'].isin(selected_players)]
        
        if len(available_players) > 0:
            available_players = available_players.reset_index(drop=True)
            selected_player = random.choices(available_players['Player'], available_players['Count'])[0]
            draft_results.append(selected_player)
            selected_players.add(selected_player)

    # Store the results of the simulation
    all_results.append(draft_results)
```

```python
simulation_results = pd.DataFrame(all_results)
```

```python
simulation_results.head()
```

```python
# Calculate the frequency/count of each player in the first round
player_counts = simulation_results.iloc[:, :32].stack().value_counts()

# Sort the players based on their count in descending order
sorted_players = player_counts.index.tolist()

# Define the colormap using a gradient color scheme
cmap = plt.get_cmap('Reds')  # Choose the desired colormap
normalize = mcolors.Normalize(vmin=0, vmax=max(player_counts))  # Normalize the data range


# Create the bar chart
plt.figure(figsize=(12, 6))
plt.bar(range(len(sorted_players)), player_counts[sorted_players], color=cmap(normalize(player_counts)))
plt.xticks(range(len(sorted_players)), sorted_players, rotation=90)
plt.xlabel('Player')
plt.ylabel('Frequency')
plt.title('Frequency of Players Taken in the First Round')
plt.tight_layout()
plt.show()

# Report the top 5 most frequently drafted players
top_players = player_counts.head(10)
print('Top 10 Most Frequently Drafted Players in the First Round:')
print(top_players)
```

```python
# Investigating a Steeler's favorite, Joey Porter Jr. 

# Filter the dataset for Joey Porter Jr.
joey_porter_df = simulation_results[simulation_results == 'Joey Porter Jr.']

# Count the frequency of each pick number
pick_counts = joey_porter_df.count()

# Count the number of simulations where Joey Porter Jr. was not picked before pick 17
count_not_picked_before_17 = 0

for _, row in simulation_results.iterrows():
    if 'Joey Porter Jr.' not in row.values[:16]:
        count_not_picked_before_17 += 1

# Calculate the probability
probability = count_not_picked_before_17 / len(simulation_results)

# Create the bar chart
fig, ax = plt.subplots()

# Set the color scheme
colors = ['#000000', '#FFB612']  # Black and gold colors

# Plot the bars with black and gold colors
ax.bar(pick_counts.index, pick_counts.values, color='#FFB612')

# Add a line plot for the pick number at which Joey Porter Jr. was selected
ax.plot(pick_counts.index, pick_counts.values, color='black', marker='o', linestyle='-')

# Set the chart title and labels
ax.set_title('Joey Porter Jr. Selections in 1,000 Simulations', color='#000000')
ax.set_xlabel('Pick Number', color='#000000')
ax.set_ylabel('Frequency', color='#000000')

# Add the probability annotation in the upper right-hand corner
annotation_text = f"The probability that Joey\n Porter Jr. will be available\n for the Steelers is: {probability:.2%}."
bbox_props = dict(boxstyle='square', facecolor='white', edgecolor='black')
plt.text(0.6, 0.9, annotation_text, fontsize=9, ha='left', va='center',
         transform=plt.gca().transAxes, bbox=bbox_props, wrap=True)

# Set the tick colors
ax.tick_params(colors='#000000')

# Show the chart
plt.show()





# Print the probability
print(f"The probability that Joey Porter Jr. will be available for the Steelers is: {probability:.2%}")
```
